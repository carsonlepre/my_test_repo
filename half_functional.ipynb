```python
import csv
import os
import pandas as pd
```

```python
unplugged_vehicle_id = []
unplugged_vehicle_location = []
unplugged_vehicle_access_group = []
unplugged_vehicle = []
```

```python
csvpath = os.path.join("scootersnoheader.csv")
```

```python
file = "scootersnoheader.csv"
glance = pd.read_csv(file)
glance.head(15)
```

```python
with open(csvpath, newline="") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")

    for row in csvreader:
        id = row[0]
        garage = row[8]
        charging = row[10]
        access_group = row[4]
        currently_in_garage = row[8]
        
        if int(id) >= 450 and garage == "TRUE" and charging == "FALSE":
            unplugged_vehicle.append([row[0], row[7], row[4]])
            print("Vehicle number " + str(id) + " needs to be plugged in at " + str(garage) + ". " "It is in " + str(access_group) + " access group.")
```

```python
with open("Unplugged_Log.md", "w") as output:
    for i in range(len(unplugged_vehicle_id)):
        output.write(f"Vehicle number {unplugged_vehicle_id[i]} was unplugged at {unplugged_vehicle_location[i]}. It was in {unplugged_vehicle_access_group[i]} access group."'\r')
```
